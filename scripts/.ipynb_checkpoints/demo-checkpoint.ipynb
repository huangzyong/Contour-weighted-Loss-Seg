```python
# 查看数据信息

import os
import numpy as np
import SimpleITK as sitk

data_dir = '/home/hzy/projects/PENGWIN/data/data-resample/task1'
file_list = os.listdir(data_dir)
print("total num: ", len(file_list))
for file in file_list:
    path_file = os.path.join(data_dir, file, 'label.nii.gz')
#     print(path_file)
    data = sitk.ReadImage(path_file)
    space = data.GetSpacing()
    size = data.GetSize()
    print("file: {}, size: {}, space: {}".format(file, size, space))
```
