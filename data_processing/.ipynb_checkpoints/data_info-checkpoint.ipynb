```python
# 查看数据信息，spacing, size， number of label 

import SimpleITK as sitk 
import matplotlib.pyplot as plt
import numpy as np

data_img = '/home/hzy/projects/PENGWIN/data/data-ori/task1/PENGWIN_CT_train_images/005.mha'
data_tar = '/home/hzy/projects/PENGWIN/data/data-ori/task1/PENGWIN_CT_train_labels/005.mha'
f_img = sitk.ReadImage(data_tar)
np_tar = sitk.GetArrayFromImage(f_img)
print("type: ", np_img.dtype)
spacing = f_img.GetSpacing()
size = f_img.GetSize()
print("spacing: ", spacing)
print("size: ", size)
count = 0
for i in range(1,31):
    u = (np_tar==i).sum()
    if u!=0:
        print("cls: {}, sum: {}".format(i, u))
        count += 1
print("num: ", count)

np_img = sitk.GetArrayFromImage(sitk.ReadImage(data_img))
print(np_img.shape)

num=150
plt.figure()
plt.subplot(121)
plt.imshow(np_img[num, :, :], cmap='gray')
plt.subplot(122)
plt.imshow(np_tar[num, :, :], cmap='gray')
plt.show()

```

```python
# 统计数据信息，标签分布信息
import os
import numpy as np
import SimpleITK as sitk
import matplotlib.pyplot as plt

testFolder = '/home/hzy/Projects/PENGWIN/data1/data-resample/task1'
all_paths = []
test_list = sorted(os.listdir(testFolder))

# 统计类别不平衡
cls_list = []
for item in test_list:
    # print(item)
    f_path = os.path.join(testFolder, item, 'label.nii.gz')

    f_img = sitk.ReadImage(f_path)
    np_tar = sitk.GetArrayFromImage(f_img)
    # print("np_tar: ", np_tar.shape)
    num = []
    for i in range(1,31):
#         u = (np_tar==i).sum()
        if i in np_tar:
            num.append(i)
            cls_list.append(i)
    print("num: ", item, num)

# print(cls_list)
for i in range(1,31):
    num_c = cls_list.count(i)
    print("num {}: {}".format(i, num_c))
```
