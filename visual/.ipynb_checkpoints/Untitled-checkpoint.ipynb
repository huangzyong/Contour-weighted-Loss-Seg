```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage  

# 生成一个200x200的二值图像
label = np.zeros((200, 200), dtype=np.uint8)

# 中间120x120的位置设置为1
label[50:150, 50:150] = 1

# 使用形态学腐蚀操作
structuring_element = np.ones((3, 3), dtype=np.uint8)
eroded_image = ndimage.binary_erosion(label, structure=structuring_element, iterations=10)

# 获得轮廓
contour_image = label - eroded_image

# 可视化
plt.figure(figsize=(8,4))
fig, axes = plt.subplots(1, 3, figsize=(12, 4))
axes[0].imshow(label, cmap='gray')
axes[0].set_title('Original Image')
axes[1].imshow(eroded_image, cmap='gray')
axes[1].set_title('Eroded Image')
axes[2].imshow(contour_image, cmap='gray')
axes[2].set_title('Contour Image')

for ax in axes:
    ax.axis('off')
plt.tight_layout()
plt.show()


# 预测结果
pred = np.zeros((200, 200), dtype=np.uint8)
pred[45:160, 40:180] = 1
idx_contour = np.where(contour_image==1)
pred_contour = np.zeros((200, 200), dtype=np.uint8)
pred_contour[idx_contour] = pred[idx_contour]
print(pred_contour.shape)

idx_nocontour = np.where(contour_image!=1)
pred_nocontour = np.zeros((200, 200), dtype=np.uint8)
pred_nocontour[idx_nocontour] = pred[idx_nocontour]

print(pred_nocontour.shape)

# 可视化
plt.figure(figsize=(8,4))
fig, axes = plt.subplots(1, 3, figsize=(12, 4))
axes[0].imshow(pred, cmap='gray')
axes[0].set_title('Original Image')
axes[1].imshow(pred_contour, cmap='gray')
axes[1].set_title('Eroded Image')
axes[2].imshow(pred_nocontour, cmap='gray')
axes[2].set_title('Contour Image')

for ax in axes:
    ax.axis('off')
plt.tight_layout()
plt.show()

dice_total = 2 * (label*pred).sum() / (label.sum() + pred.sum()+1e-3)
dice_contour = 2 * (contour_image*pred_contour).sum() / (contour_image.sum() + pred_contour.sum()+1e-3)
dice_nocontour = 2 * (eroded_image*pred_nocontour).sum() / (eroded_image.sum() + pred_nocontour.sum()+1e-3)
print("dice_total: ", dice_total)
print('dice_contour', dice_contour)
print("dice_nocontour", dice_nocontour)

print("loss_total: ", 1-dice_total)
print('loss_contour', 1-dice_contour)
print("loss_nocontour", 1-dice_nocontour)

# 预测结果
pred2 = np.zeros((200, 200), dtype=np.uint8)
pred[30:100, 70:140] = 1
idx_contour = np.where(contour_image==1)
pred_contour = np.zeros((200, 200), dtype=np.uint8)
pred_contour[idx_contour] = pred2[idx_contour]
print(pred_contour.shape)

idx_nocontour = np.where(contour_image!=1)
pred_nocontour = np.zeros((200, 200), dtype=np.uint8)
pred_nocontour[idx_nocontour] = pred2[idx_nocontour]

print(pred_nocontour.shape)

# 可视化
plt.figure(figsize=(8,4))
fig, axes = plt.subplots(1, 3, figsize=(12, 4))
axes[0].imshow(pred2, cmap='gray')
axes[0].set_title('Original Image')
axes[1].imshow(pred_contour, cmap='gray')
axes[1].set_title('Eroded Image')
axes[2].imshow(pred_nocontour, cmap='gray')
axes[2].set_title('Contour Image')

for ax in axes:
    ax.axis('off')
plt.tight_layout()
plt.show()

dice_total = 2 * (label*pred).sum() / (label.sum() + pred.sum()+1e-3)
dice_contour = 2 * (contour_image*pred_contour).sum() / (contour_image.sum() + pred_contour.sum()+1e-3)
dice_nocontour = 2 * (eroded_image*pred_nocontour).sum() / (eroded_image.sum() + pred_nocontour.sum()+1e-3)
print("dice_total: ", dice_total)
print('dice_contour', dice_contour)
print("dice_nocontour", dice_nocontour)

print("loss_total: ", 1-dice_total)
print('loss_contour', 1-dice_contour)
print("loss_nocontour", 1-dice_nocontour)

```
